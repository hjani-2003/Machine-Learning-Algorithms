```python
from sklearn.preprocessing import StandardScaler
```

```python
cols = features.columns
features = StandardScaler().fit_transform(features)
features = pd.DataFrame(features, columns=[cols])
```

```python
features
```

```python
# Visual Python: Machine Learning > Fit/Predict
model.fit(features)
```

```python
# Visual Python: Machine Learning > Dimension
from sklearn.decomposition import PCA

# model = PCA(n_components=4, random_state=4)
model = PCA(random_state=4)
trans = model.fit_transform(features)
```

```python

trans
```

```python
pca_df = pd.DataFrame(data=trans, columns=['PC1','PC2', 'PC3','PC4','PC5','PC6','PC7','PC8','PC9'])
```

```python
pca_df
```

```python
eigenvalues = model.explained_variance_
prop_var = eigenvalues / np.sum(eigenvalues)
plt.figure(figsize=(14,10))
plt.plot(np.arange(1, len(eigenvalues)+1), 
         eigenvalues, marker='o')
plt.xlabel('Principal Component',
           size = 20)
plt.ylabel('Eigenvalue',
           size = 20)
plt.title('Figure 2: Scree Plot for Eigenvalues',
          size = 25)
plt.axhline(y=1, color='r', 
            linestyle='--')
plt.grid(True)
```

```python
from sklearn.decomposition import PCA

model = PCA(n_components=2, random_state=4)
trans = model.fit_transform(features)
pca_df = pd.DataFrame(data=trans, columns=['PC1','PC2'])
df = pd.DataFrame(model.components_, columns=list(features.columns))
```
